```python
# Function to generate the LaTeX string for the qubit state with theta and phi in degrees
def generate_latex(theta_deg, phi_deg):
    # Convert degrees to radians
    theta_value = theta_deg * pi / 180
    phi_value = phi_deg * pi / 180

    # Calculate the values inside the cos and sin functions
    cos_term = cos(theta_value / 2)
    sin_term = sin(theta_value / 2)

    # Convert the degree values to LaTeX string with degree symbol
    theta_latex = f"{theta_deg}^\\circ"
    phi_latex = f"{phi_deg}^\\circ"

    # Construct the LaTeX string dynamically
    latex_output = rf"|\psi\rangle_{{\mathrm{{initial}}}} = \cos\left(\frac{{{theta_latex}}}{{2}}\right) |0\rangle + e^{{i{phi_latex}}} \sin\left(\frac{{{theta_latex}}}{{2}}\right) |1\rangle"

    # Display the LaTeX equation
    display(Math(latex_output))

# Set specific values for theta and phi in degrees
theta_deg = 45  # Example: 45 degrees
phi_deg = 90    # Example: 90 degrees

# Call the function to generate and display LaTeX
generate_latex(theta_deg, phi_deg)
```

# **Qubit state on Bloch Sphere**
$$|\psi\rangle=\cos\frac{\theta}{2} |0\rangle + e^{i \phi}\sin\frac{\theta}{2}|1\rangle$$

```python
theta = widgets.FloatSlider( # in the unit of degree
    value=0,
    min=0,
    max=180,
    step=1,
    description= 'theta',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
phi = widgets.FloatSlider( # in the unit of degree
    value=0,
    min=0,
    max=360,
    step=1,
    description= 'phi',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)

def visualize_bloch(theta, phi):
    theta_rad = theta*np.pi/180 # rescaling to radian
    phi_rad = phi*np.pi/180 # rescaling to radian
    # define the initial state
    psi = np.cos(theta_rad/2)*basis(2,0) + np.exp(1j*phi_rad)*np.sin(theta_rad/2)*basis(2,1)
    b = Bloch()
    b.make_sphere()
    b.add_states(psi)
    b.render()
    b.show()

    # Call the function to generate and display LaTeX
    generate_latex(theta, phi)

out = widgets.interactive_output(visualize_bloch, {'theta': theta, 'phi': phi})

widgets.HBox([widgets.VBox([theta, phi]), out])
```

# **Apply Pauli operators**
\begin{align}
I &\equiv
\begin{bmatrix}
1 & 0 \\
0 & 1 \\
\end{bmatrix}
\qquad
X \equiv
\begin{bmatrix}
0 & 1 \\
1 & 0 \\
\end{bmatrix}\\
Y &\equiv
\begin{bmatrix}
0 & -i \\
i & 0 \\
\end{bmatrix}
\quad
Z \equiv
\begin{bmatrix}
1 & 0 \\
0 & -1 \\
\end{bmatrix}
\end{align}

The initial state is green and the 2nd state is orange vector.

```python
theta = widgets.FloatSlider( # in degree scale
    value=0,
    min=0,
    max=180,
    step=1,
    description= 'theta',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
phi = widgets.FloatSlider( # in degree scale
    value=0,
    min=0,
    max=360,
    step=1,
    description= 'phi',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
pauli_ope = widgets.Dropdown(
    options=['x', 'y', 'z', 'I'],
    value='I',
    #rows=10,
    description='Pauli Gate',
    disabled=False
)
ope_dict = {"x":sigmax(), "y":sigmay(),"z":sigmaz(),"I":qeye(2),}
def visualize_bloch(theta, phi, pauli_ope):
    theta_rad = theta*np.pi/180
    phi_rad = phi*np.pi/180
    # define the initial state
    psi = np.cos(theta_rad/2)*basis(2,0) + np.exp(1j*phi_rad)*np.sin(theta_rad/2)*basis(2,1)

    # apply an operator
    psi2 = ope_dict[pauli_ope]*psi

    # plot on Bloch Sphere
    b = Bloch()
    b.make_sphere()
    b.add_states(psi)
    b.add_states(psi2)
    b.render()
    b.show()

    # Call the function to generate and display LaTeX
    generate_latex(theta, phi)

out = widgets.interactive_output(visualize_bloch, {'theta': theta, 'phi': phi, "pauli_ope" : pauli_ope})

widgets.HBox([widgets.VBox([theta, phi, pauli_ope]), out])
```

# **Apply rotation operators**
\begin{align}
  R_x(\theta) &\equiv
  e^\frac{-i\theta X}{2} = \cos\left(\frac{\theta}{2}\right)\, I - i \sin\left(\frac{\theta}{2}\right)\, X=
  \begin{bmatrix}
    \cos\left(\frac{\theta}{2}\right) & -i\sin\left(\frac{\theta}{2}\right) \\
    -i\sin\left(\frac{\theta}{2}\right) & \cos\left(\frac{\theta}{2}\right) \\
  \end{bmatrix} \\
  R_y(\theta) &\equiv
  e^\frac{-i\theta Y}{2} = \cos\left(\frac{\theta}{2}\right)\, I - i \sin\left(\frac{\theta}{2}\right)\, Y=
  \begin{bmatrix}
    \cos\left(\frac{\theta}{2}\right) & -\sin\left(\frac{\theta}{2}\right) \\
    \sin\left(\frac{\theta}{2}\right) & \cos\left(\frac{\theta}{2}\right) \\
  \end{bmatrix} \\
  R_z(\theta) &\equiv
  e^\frac{-i\theta Z}{2} = \cos\left(\frac{\theta}{2}\right)\, I - i \sin\left(\frac{\theta}{2}\right)\, Z=
  \begin{bmatrix}
    e^{-i\frac{\theta}{2}} & 0 \\
    0 & e^{i\frac{\theta}{2}} \\
  \end{bmatrix}
\end{align}

The initial state is green and the 2nd state is orange vector.

```python
# definintion of the rotation operators
def rotate_x(theta):
    return np.cos(theta/2)*qeye(2) - 1j*np.sin(theta/2)*sigmax()
def rotate_y(theta):
    return np.cos(theta/2)*qeye(2) - 1j*np.sin(theta/2)*sigmay()
def rotate_z(theta):
    return np.cos(theta/2)*qeye(2) - 1j*np.sin(theta/2)*sigmaz()

theta = widgets.FloatSlider( # degree scale
    value=0,
    min=0,
    max=180,
    step=1,
    description= 'theta',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
phi = widgets.FloatSlider( # degree scale
    value=0,
    min=0,
    max=360,
    step=1,
    description= 'phi',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
pauli_ope = widgets.Dropdown(
    options=['Rx', 'Ry', 'Rz', 'I'],
    value='I',
    #rows=10,
    description='Rotation Gate',
    disabled=False
)
rotation_angle = widgets.FloatSlider(
    value=0,
    min=0,
    max=360,
    step=1,
    description= 'rotation angle',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
def visualize_bloch(theta, phi, pauli_ope, rotation_angle):
    theta_rad = theta*np.pi/180
    phi_rad = phi*np.pi/180
    rotation_angle = rotation_angle*np.pi/180

    ope_dict = {"Rx":rotate_x(rotation_angle), "Ry":rotate_y(rotation_angle),"Rz":rotate_z(rotation_angle),"I":qeye(2),}

    # define the initial state
    psi = np.cos(theta_rad/2)*basis(2,0) + np.exp(1j*phi_rad)*np.sin(theta_rad/2)*basis(2,1)

    # apply an operator
    psi2 = ope_dict[pauli_ope]*psi

    # plot on Bloch Sphere
    b = Bloch()
    b.make_sphere()
    b.add_states(psi)
    b.add_states(psi2)
    b.render()
    b.show()

    # Call the function to generate and display LaTeX
    generate_latex(theta, phi)

out = widgets.interactive_output(visualize_bloch, {'theta': theta, 'phi': phi, "pauli_ope" : pauli_ope, "rotation_angle":rotation_angle})

widgets.HBox([widgets.VBox([theta, phi, pauli_ope, rotation_angle]), out])
```

Apply Clifford
\begin{align}
H &\equiv
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & 1 \\
1 & 1 \\
\end{bmatrix}\\
S &\equiv
\begin{bmatrix}
1 & 0 \\
0 & i \\
\end{bmatrix}\\
T &\equiv
\begin{bmatrix}
1 & 0 \\
0 & e^{i\pi/4} \\
\end{bmatrix}
\end{align}


```python
# definintion of the operators
def hadamard():
    matrix = np.array([
        [1, 1],
        [1, -1]
    ])
    return Qobj(1/np.sqrt(2)*matrix)
def phase():
    matrix = np.array([
        [1, 0],
        [0, 1j]
    ])
    return Qobj(matrix)
def pi_8():
    matrix = np.array([
        [1, 0],
        [0, np.exp(1j*np.pi/4)]
    ])
    return Qobj(matrix)

theta = widgets.FloatSlider( # degree scale
    value=0,
    min=0,
    max=180,
    step=1,
    description= 'theta',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
phi = widgets.FloatSlider( # degree scale
    value=0,
    min=0,
    max=360,
    step=1,
    description= 'phi',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)
ope = widgets.Dropdown(
    options=['H', 'S', 'T', 'I'],
    value='I',
    #rows=10,
    description='Gate',
    disabled=False
)

def visualize_bloch(theta, phi, ope):
    ope_dict = {"H":hadamard(), "S":phase(),"T":pi_8(),"I":qeye(2),}

    theta_rad = theta*np.pi/180
    phi_rad = phi*np.pi/180

    # define the initial state
    psi = np.cos(theta_rad/2)*basis(2,0) + np.exp(1j*phi_rad)*np.sin(theta_rad/2)*basis(2,1)

    # apply an operator
    psi2 = ope_dict[ope]*psi

    # plot on Bloch Sphere
    b = Bloch()
    b.make_sphere()
    b.add_states(psi)
    b.add_states(psi2)
    b.render()
    b.show()

    # Call the function to generate and display LaTeX
    generate_latex(theta, phi)

out = widgets.interactive_output(visualize_bloch, {'theta': theta, 'phi': phi, "ope" : ope})

widgets.HBox([widgets.VBox([theta, phi, ope]), out])
```
